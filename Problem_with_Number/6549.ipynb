```python
# https://www.acmicpc.net/problem/6549

"""
시간초과
좌측 우측 구한 후, 가운데 부분 구하기
이때 가운데 탐색 알고리즘 n으로 바텀업 방식 (가장 낮은 블럭부터 탐색 시작)

시간초과 원인 : 정렬 알고리즘 (nlogn)이 끼어있어서 그런 것 같다. 정렬을 사용하지 않고도 가능하므로 수정해보자
"""

import sys

def volume(nums, n):
    if n == 1:
        return nums[0]
    else:        
        flag = n//2
        left = volume(nums[:flag], flag)
        right = volume(nums[flag:], n - flag)

        middle_min = min(nums[flag-1], nums[n-flag-1])
        middle_idx = flag-1 if nums[flag-1] < nums[n-flag-1] else n-flag-1
        nums2 = sorted([[nums[i],i] for i in range(n)])

        # print("@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@")
        # print("nums:",nums)
        # print("middle: ",middle_min, middle_idx)

        l_idx = 0
        r_idx = n-1
        result = 0
        for i in range(n):
            # print("v, idx: ",nums2[i])
            [v, idx] = nums2[i]
            if idx == middle_idx:
                l_idx = l_idx+1 if nums[l_idx] < v else l_idx
                r_idx = r_idx-1 if nums[r_idx] < v else r_idx
                # print("l_idx, r_idx, score: ",l_idx, r_idx, v*(r_idx - l_idx + 1))
                result = max(result, v*(r_idx - l_idx + 1))
                break
            elif idx < middle_idx:
                if idx > l_idx:
                    l_idx = idx
                    result = max(result, v*(r_idx - l_idx + 1)) if v == nums[r_idx] else max(result, v*(r_idx - l_idx))
            elif idx > middle_idx:
                if idx < r_idx:
                    r_idx = idx
                    result = max(result, v*(r_idx - l_idx + 1)) if v == nums[l_idx] else max(result, v*(r_idx - l_idx))

        # print("left, merge, right: ",left, result, right)
        # print("--------------------------------------")
        return max(result, left, right)

while True:
    nums = list(map(int, sys.stdin.readline().split()))
    if nums[0] == 0:
        break 
    print(volume(nums[1:], nums[0]))
```
