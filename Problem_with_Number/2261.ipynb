```python
# https://www.acmicpc.net/problem/2261

import sys

def shortest_point(points, n):
    if n == 2:
        return (points[0][0] - points[1][0])**2 + (points[0][1] - points[1][1])**2 # 0:x 1:y
    elif n == 3:
        result = 1e9
        for i in range(2):
            for j in range(i+1,3):
                result = min(result, (points[i][0] - points[j][0])**2 + (points[i][1] - points[j][1])**2)
        return result
    else:
        flag = n//2
        left = points[:flag]
        right = points[flag:]

        # print(points)
        # print(left, n//2)
        # print(right, n-n//2)

        left_result = shortest_point(left, n//2)
        right_result = shortest_point(right, n - n//2)
        result = (min(left_result, right_result))
        d = result**0.5

        # print(left_result, right_result, d)

        middlex = (points[flag-1][0] + points[flag][0]) / 2
        lx = middlex - d
        rx = middlex + d
        
        # print(middlex, lx, rx)

        start = end = -1
        for i in range(n):
            start = i if start == -1 and points[i][0] >= lx else start
            end = i if points[i][0] <= rx else end
        middlep = sorted(points[start:end+1], key = lambda x : x[1])
        middlen = end - start + 1

        # print(start, end, middlen)

        for i in range(middlen):
            for j in range(1, 4):
                if i-j >= 0:
                    result = min(result, (middlep[i][0] - middlep[i-j][0])**2 + (middlep[i][1] - middlep[i-j][1])**2)
                if i+j < middlen:
                    result = min(result, (middlep[i][0] - middlep[i+j][0])**2 + (middlep[i][1] - middlep[i+j][1])**2)
        
        return result

n = int(input())
points = sorted([list(map(int,sys.stdin.readline().split())) for _ in range(n)])

print(shortest_point(points, n))
```
