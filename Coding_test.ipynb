```python
# https://www.acmicpc.net/problem/5073
while True:
    a, b, c = sorted(map(int, input().split()))
    if a == b == c == 0:
        break
    if a + b <= c:
        print('Invalid')
    elif a == b == c:
        print('Equilateral')
    elif a == b or b == c:
        print('Isosceles')
    else:
        print('Scalene')
```

```python
# https://www.acmicpc.net/problem/10431
T = int(input())
for _ in range(T):
    p, *a = map(int, input().split())
    
    i = 0
    cnt = 0
    now_max = 0
    while i < 20:
        if now_max > a[i]:
            for j in range(i):
                if a[j] > a[i]:
                    cnt += len(a[j:i])
                    a = a[:j] + [a[i]] + a[j:i] + a[i+1:]
                    i = 0
                    now_max = 0
                    break
        else:
            now_max = a[i]
            i += 1
    print(p, cnt)    
```

```python
# https://www.acmicpc.net/problem/8979

T, K = map(int, input().split())
ex = []
for _ in range(T):
    p, x1, x2, x3 = map(int, input().split())
    ex.append([x1, x2, x3, p])
ex = list(sorted(ex, key=lambda x: (x[0], x[1], x[2]), reverse=True))

last_ex = ex[0]
last_score = 1
for i in range(T):
    _, _, _, p = ex[i]
    if last_ex[:-1] == ex[i][:-1]:
        now_score = last_score
    else:
        now_score = i + 1
        last_ex = ex[i]
        last_score = now_score
    if K == p:
        print(now_score)
        break        
```

```python
# https://www.acmicpc.net/problem/4659

while True:
    word = input()
    if word == 'end':
        break

    acceptable = True
    for i in range(len(word) - 1):
        if word[i] == word[i+1] and word[i] not in ['e', 'o']:
            acceptable = False
            break
    for i in range(len(word) - 2):
        triple_1 = True # ['a', 'e', 'i', 'o', 'u']
        triple_2 = True # not ['a', 'e', 'i', 'o', 'u']
        for j in range(3):
            triple_1 = triple_1 & (word[i+j] in ['a', 'e', 'i', 'o', 'u'])
            triple_2 = triple_2 & (word[i+j] not in ['a', 'e', 'i', 'o', 'u'])
        if triple_1 or triple_2:
            acceptable = False
    is___there = False
    for i in range(len(word)):
        if word[i] in ['a', 'e', 'i', 'o', 'u']:
            is___there = True
    acceptable = acceptable and is___there
    
    if acceptable:
        print(f"<{word}> is acceptable.")
    else:
        print(f"<{word}> is not acceptable.")
```

```python
# https://www.acmicpc.net/problem/25757

N, P = input().split()
N = int(N)
P = {'Y':1, 'F':2, "O":3}[P]

names = [input() for _ in range(N)]
played = {}
for name in names:
    played[name] = False

plays = 0
queue = 0
for i in range(N):
    name = names[i]
    if played[name] == False:
        queue += 1
        played[name] = True
    if queue == P:
        queue = 0
        plays += 1
print(plays)
```

```python
# https://www.acmicpc.net/problem/20125

N = int(input())
now_map = [list(input()) for _ in range(N)] # [y][x]

head_x = -1
heart = None
left_arm = 0
right_arm = 0
line = 0
left_leg = 0
right_leg = 0
for y in range(N):
    for x in range(N):
        if head_x == -1 and now_map[y][x] == "*":
            head_x = x
            heart = [y+1, x]
        if head_x != -1:
            if y == heart[0] and x < head_x and now_map[y][x] == "*":
                left_arm += 1
            elif y == heart[0] and x > head_x and now_map[y][x] == "*":
                right_arm += 1
            elif y > heart[0] and x == head_x and now_map[y][x] == "*":
                line += 1
            elif y > heart[0] and x == head_x-1 and now_map[y][x] == "*":
                left_leg += 1
            elif y > heart[0] and x == head_x+1 and now_map[y][x] == "*":
                right_leg += 1

print(f"{heart[0]+1} {heart[1]+1}")
print(left_arm, right_arm, line, left_leg, right_leg)
```

```python
# https://www.acmicpc.net/problem/1205

N, score, P = map(int, input().split())

if N == 0:
    print(1)
else:
    ranks = list(map(int, input().split()))
    if N == P and ranks[-1] >= score:
        print(-1)
    else:
        rank = N+1
        for i in range(N):
            if score >= ranks[i]:
                rank = i+1
                break
        print(rank)
```

```python
# https://www.acmicpc.net/problem/1244
N = int(input())
switches = [-1] + list(map(int, input().split()))
num_s = int(input())

for _ in range(num_s):
    gender, number = map(int,input().split())
    if gender == 1:
        idx = number
        while idx <= N:
            switches[idx] = 1 - switches[idx]
            idx += number
    else:
        idx = number
        switches[idx] = 1 - switches[idx]
        i = 1
        while ((idx - i) >= 1) and ((idx + 1) <= len(switches)):
            if idx - i < 1 or idx + i > N: break
            if switches[idx-i] == switches[idx+i]:
                switches[idx-i] = 1 - switches[idx-i]
                switches[idx+i] = 1 - switches[idx+i]
                i += 1
            else:
                break

result = ""
switches = switches[1:]
for i in range(len(switches)):
    result += f"{switches[i]} "
    if (i+1) % 20 == 0:
        print(result.rstrip())
        result = ""
if result != "":
    print(result.rstrip())
```

```python
# https://www.acmicpc.net/problem/9017

T = int(input())
for _ in range(T):
    length = int(input())
    scores = list(map(int, input().split()))
    valid = {}
    for s in scores:
        if s not in valid.keys(): valid[s] = 1
        else: valid[s] += 1
    for key in valid.keys():
        if valid[key] < 6: valid[key] = False
        else: valid[key] = True

    dicts = {}
    now_score = 1
    for i in range(length):
        key = scores[i]
        if valid[key] == True:
            if key not in dicts.keys():
                dicts[key] = [1, now_score, -1]
            else:
                if dicts[key][0] == 4:
                    dicts[key][2] = now_score
                elif dicts[key][0] <= 4:
                    dicts[key][1] += now_score
                dicts[key][0] += 1

            now_score += 1

    min_score = 9999999999999
    fifth_time = -1
    winner = -1
    for key in dicts.keys():
        if dicts[key][0] == 6:
            if dicts[key][1] < min_score:
                min_score = dicts[key][1]
                winner = key
                fifth_time = dicts[key][2]
            elif dicts[key][1] == min_score and dicts[key][2] < fifth_time:
                winner = key
                fifth_time = dicts[key][2]
    print(winner)
```

```python
# https://www.acmicpc.net/problem/17266

N = int(input())
M = int(input())
X = list(map(int, input().split()))

height = 0
if M > 1:
    distances = max([X[i+1] - X[i] for i in range(len(X)-1)])
    height = distances // 2 + (1 if distances % 2 == 1 else 0)
if X[0] != 0:
    height = max(height, X[0])
if X[-1] != N:
    height = max(height, N - X[-1])

print(height)
```

```python
# https://www.acmicpc.net/problem/20920
import sys

N, M = map(int, sys.stdin.readline().rstrip().split())
dicts = {}
words = [sys.stdin.readline().rstrip() for _ in range(N)]
for word in words:
    if len(word) >= M:
        dicts[word] = [0, len(word)]

for word in words:
    if len(word) >= M:
        dicts[word][0] += 1

infos = []
for key in dicts.keys():
    infos.append([-dicts[key][0], -dicts[key][1], key])
infos = sorted(infos, key=lambda x: (x[0],x[1],x[2]))

for info in infos:
    print(info[2])
```

```python
# https://www.acmicpc.net/problem/2512

N = int(input())
demands = list(map(int, input().split()))
budget = int(input())

if sum(demands) < budget:
    print(max(demands))
else:
    minv = 0
    maxv = max(demands)
    optimal = -1
    while minv <= maxv:
        nowv = (maxv + minv) // 2
        nowd = sum([min(d, nowv) for d in demands])
        if nowd <= budget:
            optimal = nowv
            minv = nowv + 1
        else:
            maxv = nowv - 1
    print(optimal)
```

```python
# https://www.acmicpc.net/problem/21921

N, X = map(int, input().split())
logs = list(map(int, input().split()))

base = sum(logs[:X])
max_log = base
max_cnt = 1
for i in range(1, N-(X-1)):
    base -= logs[i-1]
    base += logs[i+X-1]
    if base == max_log:
        max_cnt += 1
    elif base > max_log:
        max_log = base
        max_cnt = 1

if max_log == 0:
    print("SAD")
else:
    print(max_log)
    print(max_cnt)
```
